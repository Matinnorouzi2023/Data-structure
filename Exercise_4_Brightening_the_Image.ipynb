# کد پایتون برای روش تصحیح گاما

# توضیحات کد:
تصویر با استفاده از cv2.imread بارگذاری می‌شود.
کاربر مقدار gamma را وارد می‌کند. برای تصویر تاریک شما، مقادیر پیشنهادی می‌تواند gamma=0.4 باشد (مانند مثال‌های ارائه‌شده با γ=1/3 یا γ=1/4).
یک جدول LUT (Look-Up Table) برای مقادیر 0 تا 255 ایجاد می‌شود که مقادیر پیکسل‌ها را با فرمول گاما تنظیم می‌کند.
تابع cv2.LUT برای اعمال تصحیح گاما روی تصویر استفاده می‌شود.
تصویر اصلی و تصویر تنظیم‌شده با استفاده از matplotlib نمایش داده می‌شوند.
در نهایت، تصویر خروجی ذخیره می‌شود

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# بارگذاری تصویر
image = cv2.imread('D:/Weeki_Wachee_spring_10079u.jpg')

# بررسی اینکه تصویر به درستی بارگذاری شده است
if image is None:
    print("Error: Could not load image. Please check the file path.")
    exit()

# درخواست مقدار گاما از کاربر
gamma = float(input('Enter the gamma value [0.1-5.0]: '))

# ایجاد جدول LUT (Look-Up Table) برای تصحیح گاما
lookUpTable = np.empty((1, 256), np.uint8)
for i in range(256):
    lookUpTable[0, i] = np.clip(pow(i / 255.0, gamma) * 255.0, 0, 255)

# اعمال تصحیح گاما
gamma_image = cv2.LUT(image, lookUpTable)

# نمایش تصویر اصلی و تصویر روشن‌شده
plt.figure(figsize=[15, 5])
plt.subplot(121)
plt.imshow(image[...,::-1])  # تبدیل BGR به RGB برای نمایش صحیح
plt.title("Original")
plt.axis('off')

plt.subplot(122)
plt.imshow(gamma_image[...,::-1])  # تبدیل BGR به RGB برای نمایش صحیح
plt.title("Gamma Corrected (gamma={})".format(gamma))
plt.axis('off')

plt.show()

# ذخیره تصویر خروجی (اختیاری)
cv2.imwrite("gamma_corrected_image.jpg", gamma_image)
```

# 3. مقایسه هر دو روش (Linear و Gamma Correction)
برای مقایسه هر دو روش، می‌توانیم هر دو را در کنار تصویر اصلی نمایش دهیم:

کد پایتون برای مقایسه

# توضیحات کد:
تصویر اصلی بارگذاری می‌شود.
روش خطی با مقادیر پیشنهادی alpha=1.5 و beta=50 اعمال می‌شود.
روش گاما با مقدار پیشنهادی gamma=0.4 اعمال می‌شود.
هر سه تصویر (اصلی، خطی، و گاما) کنار هم نمایش داده می‌شوند.
تصاویر خروجی ذخیره می‌شوند

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# بارگذاری تصویر
image = cv2.imread('D:/Weeki_Wachee_spring_10079u.jpg')

# بررسی اینکه تصویر به درستی بارگذاری شده است
if image is None:
    print("Error: Could not load image. Please check the file path.")
    exit()

# تنظیمات روش خطی
alpha = 1.5  # فاکتور کنتراست
beta = 50    # فاکتور روشنایی
linear_image = cv2.convertScaleAbs(image, alpha=alpha, beta=beta)

# تنظیمات روش گاما
gamma = 0.4  # مقدار گاما
lookUpTable = np.empty((1, 256), np.uint8)
for i in range(256):
    lookUpTable[0, i] = np.clip(pow(i / 255.0, gamma) * 255.0, 0, 255)
gamma_image = cv2.LUT(image, lookUpTable)

# نمایش تصاویر
plt.figure(figsize=[15, 5])
plt.subplot(131)
plt.imshow(image[...,::-1])  # تبدیل BGR به RGB
plt.title("Original")
plt.axis('off')

plt.subplot(132)
plt.imshow(linear_image[...,::-1])  # تبدیل BGR به RGB
plt.title("Linear (alpha={}, beta={})".format(alpha, beta))
plt.axis('off')

plt.subplot(133)
plt.imshow(gamma_image[...,::-1])  # تبدیل BGR به RGB
plt.title("Gamma (gamma={})".format(gamma))
plt.axis('off')

plt.show()

# ذخیره تصاویر خروجی (اختیاری)
cv2.imwrite("linear_adjusted_image.jpg", linear_image)
cv2.imwrite("gamma_corrected_image.jpg", gamma_image)
```

# تحلیل و نتیجه‌گیری
## روش خطی: این روش با افزایش یکنواخت روشنایی و کنتراست کار می‌کند. برای تصویر تاریک ما، این روش می‌تواند تصویر را روشن‌تر کند، اما ممکن است جزئیات در مناطق خیلی تاریک یا خیلی روشن به خوبی حفظ نشود.
## روش تصحیح گاما: این روش به صورت غیرخطی عمل می‌کند و برای روشن کردن تصاویر تاریک بسیار مناسب است، زیرا مناطق تاریک را بیشتر از مناطق روشن تحت تأثیر قرار می‌دهد. همان‌طور که در تصاویر ما با γ=1/3 و γ=1/4 دیده می‌شود، این روش می‌تواند جزئیات بیشتری را در مناطق تاریک نشان دهد.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# بارگذاری تصویر اصلی
image = cv2.imread('D:/Weeki_Wachee_spring_10079u.jpg')

# بررسی اینکه تصویر به درستی بارگذاری شده است
if image is None:
    print("Error: Could not load image. Please check the file path.")
    exit()

# لیست مقادیر گاما برای مقایسه
gamma_values = [2, 1, 1/2, 1/3, 1/4]  # شامل gamma=2, gamma=1 (اصلی), gamma=1/2, gamma=1/3, gamma=1/4
gamma_labels = ['γ=2', 'γ=1 (original)', 'γ=1/2', 'γ=1/3', 'γ=1/4']  # برچسب‌ها برای نمایش

# لیست برای ذخیره تصاویر
images = []

# اعمال تصحیح گاما برای هر مقدار گاما
for gamma in gamma_values:
    # ایجاد جدول LUT (Look-Up Table) برای تصحیح گاما
    lookUpTable = np.empty((1, 256), np.uint8)
    for i in range(256):
        lookUpTable[0, i] = np.clip(pow(i / 255.0, gamma) * 255.0, 0, 255)

    # اعمال تصحیح گاما روی تصویر
    gamma_image = cv2.LUT(image, lookUpTable)
    images.append(gamma_image)

# نمایش تصاویر
plt.figure(figsize=[15, 10])  # تنظیم اندازه شکل برای نمایش بهتر
for i in range(len(images)):
    plt.subplot(1, 5, i+1)  # 1 ردیف و 5 ستون برای نمایش تصاویر
    plt.imshow(images[i][...,::-1])  # تبدیل BGR به RGB برای نمایش صحیح
    plt.title(gamma_labels[i])
    plt.axis('off')

plt.tight_layout()  # تنظیم فاصله بین تصاویر
plt.show()

# ذخیره تصاویر خروجی (اختیاری)
for i, gamma in enumerate(gamma_values):
    cv2.imwrite(f"gamma_corrected_image_{gamma}.jpg", images[i])
```

# تحلیل نتایج

γ=1: این تصویر اصلی است و هیچ تغییری در آن اعمال نشده است (حالت خطی).

γ=2: تصویر تاریک‌تر می‌شود، زیرا مقادیر گاما بیشتر از 1 باعث کاهش روشنایی می‌شوند.

γ=1/2: تصویر روشن‌تر می‌شود، اما روشنایی کمتر از 𝛾=1/3 و 𝛾=1/4  است.

γ=1/3: تصویر باز هم روشن‌تر می‌شود، زیرا مقدار گاما کاهش یافته است.

γ=1/4: این روشن‌ترین تصویر است، زیرا کوچک‌ترین مقدار گاما بیشترین تأثیر را در روشن کردن مناطق تاریک دارد.
